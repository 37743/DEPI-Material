```python
import collections

genre_counter = collections.Counter()

for row in df.itertuples(index=False):
    genres = row.Genre.split(',')
    for genre in genres:
        genre_counter.update([genre.strip()])

genre_dict = dict(genre_counter)

print(genre_dict)
```
